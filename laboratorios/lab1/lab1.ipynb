# Iniciar o acesso à câmera

Obtemos o acesso à câmera. Seu sistema operacional pode solicitar permissões.

````python
cap = cv.VideoCapture(0)

if not cap.isOpened():
  print("Não foi possível acessar a câmera do dispositivo.")
  exit()
````

# Obter uma Fotografia

Neste etapa, podemos tirar uma foto da câmera executando as duas células a seguir.

A foto será salva com o `nome_foto` que você configurar na variável abaixo.

````python
nome_foto = 'foto1.png'
````

Execute a célula e pressione a tecla X para tirar uma foto. Pressione Q para encerrar. 

Não feche a janela de visualização da webcam, será re-utilizada neste notebook.

````python
while True:
  # Ler o frame da webcam
  ret, frame = cap.read()

  # Caso não tenha sido possível ler a webcam, ret não será True:
  if not ret:
    print("Não foi possível ler o frame da webcam.")
    break

  # Exibir o frame
  cv.imshow('frame', frame)

  # Aguarda o pressionar de uma tecla por 1ms
  key = cv.waitKey(1)

  # Ao pressionar X, salva a foto.
  if key == ord('x'):
    cv.imwrite(nome_foto, frame)

  # Ao pressionar Q, encerra.
  if key == ord('q'):
    break
````

**IMPORTANTE:** Não feche a janela com a imagem da câmera. Ela será reutilizada para outras fotografias ou captura de vídeo neste notebook.

# Obter um Vídeo

Neste etapa, podemos tirar gravar um vídeo da câmera executando as duas células a seguir.

O vídeo será salvo com o `nome_video` que você configurar na variável abaixo. Você pode controlar o FPS do vídeo também.

````python
nome_video = 'saida.avi'
fps = 25
````

Execute a célula para gravar:

````python
# Tempo que cada frame deve durar.
time_per_frame = 1 / fps

# Tamanho do vídeo da webcam em pixels.
width = cap.get(cv.CAP_PROP_FRAME_WIDTH)
height = cap.get(cv.CAP_PROP_FRAME_HEIGHT)

# Codec do vídeo
fourcc = cv.VideoWriter_fourcc(*'XVID')

# Stream para o arquivo onde será salvo o vídeo
out = cv.VideoWriter(nome_video, fourcc, fps, (int(width), int(height)))

while cap.isOpened():
  # Usamos o start_time para garantir que cada frame será capturado em um intervalo igual de tempo
  # Caso contrário o vídeo poderá ficar acelerado ou lento.
  start_time = time.time()

  # Ler o frame da webcam
  ret, frame = cap.read()

  # Caso não tenha sido possível ler a webcam, ret não será True:
  if not ret:
    print("Não foi possível ler o frame da webcam.")
    break

  # Salva o frame do vídeo e exibe
  out.write(frame)
  cv.imshow('frame', frame)

  # Calculamos quanto tempo já se foi desde quando começamos a capturar este frame
  passed_time = time.time() - start_time

  # Calculamos quanto tempo ainda devemos esperar para gravar o próximo frame em um intervalo igual de tempo
  wait_time = int((time_per_frame - passed_time) * 1000)

  if wait_time > 0 and cv.waitKey(wait_time) == ord('q'):
    break

out.release()
````

**IMPORTANTE:** Não feche a janela com a imagem da câmera. Ela será reutilizada para outras fotografias ou captura de vídeo neste notebook.
